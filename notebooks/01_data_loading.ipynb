```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Churn Label",
    data=df
)

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Gender",
    data=df
)

plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Gender",
    hue="Churn Label",
    data=df
)

plt.title("Churn by Gender")

plt.show()
```
